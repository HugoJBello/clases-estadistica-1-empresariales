```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
from scipy.stats import poisson

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

mus = [3,9]
cols = ['r','b']

for i in range(len(mus)):
    mu = mus[i]
    col = cols[i]
    mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')

    x = np.arange(poisson.ppf(0.001, mu),

                poisson.ppf(0.999, mu))


    ax.plot(x, poisson.pmf(x, mu), col+'o', ms=8, label='poisson $\lambda='+str(mu)+'$')
    
    ax.plot(x, poisson.pmf(x, mu), color=col)

    ax.vlines(x, 0, poisson.pmf(x, mu), colors=col, lw=5, alpha=0.35)
    
    ax.set_xlabel('x')

    ax.set_ylabel('p(x)')

    ax.legend(loc='best', frameon=False)

plt.show()
```

```python
from scipy.stats import hypergeom

import matplotlib.pyplot as plt

[M, n, N] = [20, 7, 12]

rv = hypergeom(M, n, N)

x = np.arange(0, n+1)

pmf_dogs = rv.pmf(x)

fig = plt.figure()

ax = fig.add_subplot(111)

ax.plot(x, pmf_dogs, 'bo',  ms=8,   label='hipergeom N=20, n=7, K=12')

ax.vlines(x, 0, pmf_dogs, lw=2)

ax.set_xlabel('x')

ax.set_ylabel('p(x)')

ax.legend(loc='best', frameon=False)

plt.show()
```
